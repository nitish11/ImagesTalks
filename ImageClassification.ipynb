```python
#Utility function
def url_to_image(url):
    # download the image, convert it to a NumPy array, and then read
    # it into OpenCV format
    resp = urllib.urlopen(url)
    image = np.asarray(bytearray(resp.read()), dtype="uint8")
    image = cv2.imdecode(image, cv2.IMREAD_COLOR)
    return image
```

### Loading the dataset and splitting train and validation


```python
from keras.datasets import mnist
(x_train, y_train), (x_val, y_val) = mnist.load_data()
print("x_train : ",np.shape(x_train))
print("y_train : ",np.shape(y_train))
print("x_val : ",np.shape(x_val))
print("y_val : ",np.shape(y_val))
```

### Visualization of Dataset

```python
index = 100
print("y_train[index] :", y_train[index])

np.set_printoptions(precision=None, threshold=None, edgeitems=None, linewidth=180, suppress=None)
print(x_train[index])
```

```python
#Preparing the input data for the training
x_train = x_train.reshape(60000,784)
x_val = x_val.reshape(10000,784)

y_train = keras.utils.to_categorical(y_train,10)
y_val = keras.utils.to_categorical(y_val,10)

print("x_train : ",np.shape(x_train))
print("y_train : ",np.shape(y_train))
print("x_val : ",np.shape(x_val))
print("y_val : ",np.shape(y_val))
```

## Machine Learning model : DecisionTree

```python
#Defining Algorithm
from sklearn.tree import DecisionTreeClassifier
decison_tree_classifier = DecisionTreeClassifier()

#Training
decison_tree_classifier.fit(x_train, y_train)
```

```python
#Validation of model
score = decison_tree_classifier.score(x_val, y_val)
print(score)
```

### Testing the Model

```python
# Testing the image
#test_image_url = "https://3qeqpr26caki16dnhd19sv6by6v-wpengine.netdna-ssl.com/wp-content/uploads/2019/02/sample_image-768x763.png"
#test_image = url_to_image(test_image_url)
```

```python
test_image = cv2.imread("resources/6.png")
plt.imshow(test_image)
grey_image = cv2.cvtColor(test_image, cv2.COLOR_BGR2GRAY)
img = np.invert(cv2.resize(grey_image,(28,28)))
print(img)
test_img = img.reshape((1,784))
```

```python
# Predicting the output for an image using Decision Tree model
img_class = decison_tree_classifier.predict(test_img)
classname = np.where(img_class[0]==1)
print("Class: ",classname)

probability = decison_tree_classifier.predict_proba(test_img)
print("Confidence: ",max(probability[0]))
```

## Deep Neural Network Model - I

```python
#Network Achitechture
model = Sequential()

#Input Layer
model.add(Dense(units=128,activation="relu",input_shape=(784,)))

#Middle Layers
model.add(Dense(units=64,activation="relu"))

#Output Layer
model.add(Dense(units=10,activation="softmax"))

#Training the model
#Defining Loss function and Optimizer
model.compile(optimizer=SGD(0.001),loss="categorical_crossentropy",metrics=["accuracy"])

#Running the training
model.fit(x_train,y_train,batch_size=32,epochs=5,verbose=1)
```

```python
# Save the weights
model.save("mnist-model.h5")

model.summary()
```

```python
print(784*128 + 128)
print(128*64 + 64)
print(64*10 + 10)
```

```python
#Evaluation of the trained model
accuracy = model.evaluate(x=x_val,y=y_val,batch_size=32)
print("Accuracy: ",accuracy[1])
```

### Testing Deep Neural Network Model

```python
# Predicting the output for an image using DeepNet model
img_class = model.predict_classes(test_img)
classname = img_class[0]
print("Class: ",classname)

probability = model.predict_proba(test_img)
print("Confidence: ",max(probability[0]))
```

## Deep Neural Network Training Model-II

```python
#Loading the data
(x_train_2, y_train_2), (x_val_2, y_val_2) = mnist.load_data()

#Preparing the dataset
y_train_2 = keras.utils.to_categorical(y_train_2,10)
y_val_2 = keras.utils.to_categorical(y_val_2,10)
```

```python
## Network Achitechture
model_2 = Sequential()

#Input Layer
model_2.add(Convolution1D(64,    # number of filter layers
                        3,    # dimension of kernel
                        input_shape=(28,28,), activation='relu'))

model_2.add(Flatten())

#Middle Layers
model_2.add(Dense(units=128,activation="relu"))
model_2.add(Dense(units=64,activation="relu"))

#Output Layer
model_2.add(Dense(units=10,activation="softmax"))
```

```python
#Defining Loss function and Optimizer
model_2.compile(optimizer=SGD(0.001),loss="categorical_crossentropy",metrics=["accuracy"])

#Running the training
model_2.fit(x_train_2,y_train_2,batch_size=32,epochs=10,verbose=1)
```

```python
## Evaluation of the trained model
accuracy = model_2.evaluate(x=x_val_2,y=y_val_2,batch_size=32)
print("Accuracy: ",accuracy[1])
```

### Testing trained model

```python
test_image_2 = cv2.imread("resources/6.png")

grey_image = cv2.cvtColor(test_image, cv2.COLOR_BGR2GRAY)
img = np.invert(cv2.resize(grey_image,(28,28)))
print(img)
test_img_2 = img.reshape((1,28,28))
```

```python
# Predicting the output for an image using DeepNet model
img_class = model_2.predict_classes(test_img_2)
classname = img_class[0]
print("Class: ",classname)

probability = model_2.predict_proba(test_img_2)
print("Confidence: ",max(probability[0]))
```
